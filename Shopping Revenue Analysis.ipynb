```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.impute import SimpleImputer
from math import sqrt
```

```python
def convert_to_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%m/%d/%Y')
    except:
        return pd.NaT

df['Open Date'] = df['Open Date'].apply(convert_to_date)
df = df.dropna(subset=['Open Date']) 
```

```python
# One-hot encode categorical columns
df = pd.get_dummies(df, columns=['City', 'City Group', 'Type'], drop_first=True)
```

```python
# Split the data into independent variables and the target variable (Revenue)
X = df.drop(['revenue', 'Open Date'], axis=1) 
y = df['revenue']  
```

```python
# Handle missing values (fill with mean)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
# Split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

# Random Forest Regressor

```python
# Create Random Forest Regressor model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
```

```python
# Calculate RMSE and MAE for the model
rmse = sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
```

# Gradient Boosting Regressor

```python
# Create Gradient Boosting Regressor model
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_test)
```

```python
# Calculate RMSE and MAE for Gradient Boosting
rmse_gb = sqrt(mean_squared_error(y_test, y_pred_gb))
mae_gb = mean_absolute_error(y_test, y_pred_gb)

print("Gradient Boosting Regression:")
print(f"Root Mean Squared Error (RMSE): {rmse_gb}")
print(f"Mean Absolute Error (MAE): {mae_gb}\n")
```

# Support Vector Regressor

```python
# Create Support Vector Regressor (SVR) model
svr_model = SVR(kernel='linear')
svr_model.fit(X_train, y_train)
y_pred_svr = svr_model.predict(X_test)
```

```python
# Calculate RMSE and MAE for SVR
rmse_svr = sqrt(mean_squared_error(y_test, y_pred_svr))
mae_svr = mean_absolute_error(y_test, y_pred_svr)

print("Support Vector Regression (SVR):")
print(f"Root Mean Squared Error (RMSE): {rmse_svr}")
print(f"Mean Absolute Error (MAE): {mae_svr}\n")
```

# Linear Regression

```python
# CreateLinear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear= linear_model.predict(X_test)
```

```python
# Calculate RMSE and MAE for Linear Regression
rmse_linear = sqrt(mean_squared_error(y_test, y_pred_linear))
mae_linear = mean_absolute_error(y_test, y_pred_linear)

print("Linear Regression:")
print(f"Root Mean Squared Error (RMSE): {rmse_linear}")
print(f"Mean Absolute Error (MAE): {mae_linear}\n")
```

# Visualize for all models

```python
# Create scatter plot 
plt.figure(figsize=(12, 8))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5, label='Random Forest')
plt.scatter(y_test, y_pred_gb, color='green', alpha=0.5, label='Gradient Boosting')
plt.scatter(y_test, y_pred_linear, color='purple', alpha=0.5, label='Linear Regression')
plt.scatter(y_test, y_pred_svr, color='red', alpha=0.5, label='SVR')

plt.xlabel('Actual Revenue')
plt.ylabel('Predicted Revenue')
plt.title('Actual vs. Predicted Revenue')
plt.grid(True)
plt.legend()

plt.show()
```

```python
# Visualize RMSE and MAE for all models using bar plots
models = ['Random Forest', 'Gradient Boosting', 'SVR', 'Linear Regression']
rmse_values = [rmse, rmse_gb, rmse_svr, rmse_linear]
mae_values = [mae, mae_gb, mae_svr, mae_linear]

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(models, rmse_values, color='skyblue')
plt.title('Root Mean Squared Error (RMSE) for Regression Models')
plt.xlabel('Model')
plt.ylabel('RMSE')
plt.ylim(0, max(rmse_values) + 1000)

plt.subplot(1, 2, 2)
plt.bar(models, mae_values, color='lightcoral')
plt.title('Mean Absolute Error (MAE) for Regression Models')
plt.xlabel('Model')
plt.ylabel('MAE')
plt.ylim(0, max(mae_values) + 1000)

plt.tight_layout()
plt.show()
```
